### 載入數據

```python
path = r"D:\google_drive\code\dataset\imdb_dataset"

title_basics = pd.read_csv(os_join(path ,"title.basics.tsv.gz"), sep="	")

title_ratings = pd.read_csv(r"D:\google_drive\code\dataset\imdb_dataset\title.ratings.tsv.gz", sep="	")
```

```python
#確認形狀，發現形狀不同
print(title_basics.shape) 
print(title_ratings.shape)
```

```python
#將兩個圖表以tconst做合併
result = pd.merge(title_basics, title_ratings, on='tconst')
```

```python
result.dtypes
```

```python
#將"\N"換成NAN
result[result == "\\N"] = np.NAN
```

```python
result["id"] = result.tconst.apply(lambda x: int(x.split("tt")[1]))
```

```python
result.index = result["id"]
```

```python
result = result.drop("tconst", axis = 1)
```
